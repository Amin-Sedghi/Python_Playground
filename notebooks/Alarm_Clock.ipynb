Importing necessary libraries.

To the get format of our countdown clock like this : 00:00 . We are using below two constant:

CLEAR = "\033[2J"

CLEAR_and_RETURN= "\033[H"

```python
from playsound import playsound
from IPython.display import clear_output
from IPython.display import Audio
import time


CLEAR = "\033[2J"
CLEAR_and_RETURN= "\033[H"
```

Below will be our main function representing a countdown timer:

( You can download your desire soundtrack from: https://pixabay.com/sound-effects/search/alarm%20clock/ )

```python
def alarm(seconds):
  time_elapsed = 0

  print(CLEAR)
  while time_elapsed < seconds:
    time.sleep(1)
    time_elapsed += 1

    time_left = seconds - time_elapsed
    minutes_left= time_left // 60
    seconds_left= time_left % 60

    clear_output(wait=True)
    print(f"Alarm will sound in: {minutes_left:02d}:{seconds_left:02d}")

  wn = Audio('/content/drive/MyDrive/Projects/Python Learning Projects/Alarm_clock/bedside.mp3', autoplay=True)
  display(wn)
```

Homepage for Alarm clock:

```python
print(f"Welcome to the Alarm clock\n")
minutes = int(input("How many minutes to wait: "))
seconds = int(input("How many seconds to wait: "))
print(f"--------------------------------------------------------------------------\n")
total_seconds= minutes * 60 + seconds

alarm(total_seconds)
```
