```python
def add_operation():
  account = input(f'Enter account name: ')
  user_name = input(f'Please enter your username: ')
  password = input(f'Please enter your password: ')
  with open('Accounts_information.txt' , 'a') as file:
    file.write(f'{account} | {user_name} | {password}\n')

def view_operation():
  with open('Accounts_information.txt' , 'r') as file:
    for line in file:
      data = line
      if '|' in data:
        account , User , Password = data.split('|')
        print(f'\nAccount= {account}\nUser= {User}\nPassword={Password}')
      else:
        print("Warning: Line does not contain the expected format:", data)


def main():
  print(f'Welcome to your password manager\n')

  while True:
    operation = input(f'For adding a new password enter (add), for reviewing your passwords enter (view) and for quiting please enter (quit): ').lower()
    if operation == 'quit':
      break
    if operation == 'add':
      add_operation()
    elif operation == 'view':
      view_operation()
    else:
      print(f"Invalid input")
      continue

main()

```
